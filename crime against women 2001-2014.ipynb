```python
#Listing of Indian states and union territories ranking by safety of women in the span of 2001-2014.
state_df=crime_df.groupby('STATE/UT')[crime_name].sum().sum(axis=1).sort_values()
df1=state_df.to_frame().rename(columns={0:'crime_in_state'})
df1.reset_index(inplace=True)
df1
```

```python
#This list shows that UP was most prone to women crimes in India in the period of 2001-2014 and Lakshadweep was the safest place.
```

```python
crime_in_state=list(crime_df.columns)[2:]
```

```python
#Which were the top five dangerous states for women in India
li=[crime_df.crime_in_state.sum()]*5
df2=crime_df.tail()
df2.insert(2,'crime_in_India',li)
df2.plot(x='STATE/UT',kind='bar',stacked=True,color=['blue','skyblue'])
plt.legend(bbox_to_anchor=(1.05,1.0))
plt.show()
df2
```

```python
#Which were the top ten reported rape cases in India and in which year and place they occured?
max_rape_cases = crime_df.sort_values('Rape', ascending = False).head(10)
rdf=max_rape_cases[['Year','STATE/UT','Rape']]
rdf.index=np.arange(1,11)
rdf
```

```python
#Interpetation :-We see from the analysis, the top states that reported the maximum number of rape cases, along with the year, in which they occured.
#Madhya Pradesh reported the maximum rape cases in the year 2014.
```

```python
#Which were the top ten reported cases of dowry deaths in India and in which year and place they occured?
max_dowry_death_cases = crime_df.sort_values('Dowry Deaths', ascending = False).head(10)
ddf=max_dowry_death_cases[['Year','STATE/UT','Dowry Deaths']]
ddf.index=np.arange(1,11)
ddf
```

```python
#Interpetation :-From the analysis, we observe that the highest reported dowry death was in Uttar Pradesh in 2014, with number of reports being 2469.
#One more point to be observed as well is that, Uttar Pradesh is the ONLY state that appears in this list.
```

```python
#Which were the top ten reported cases of domestic violence in India and in which year and place they occured?
max_Cruelty by Husband_his Relatives_cases = crime_df.sort_values('Cruelty by Husband_his Relatives', ascending = False).head(10)
vdf=max_Cruelty by Husband_his Relatives_cases[['Year','STATE/UT','Cruelty by Husband_his Relatives']]
vdf.index=np.arange(1,11)
vdf
```

```python
# were the top ten reported cases of importation of women in India and in which year and place they occured?
max_importation_case = crime_df.sort_values('Importation of Girls', ascending = False).head(10)
idf=max_importation_case[['Year','STATE/UT','Importation of Girls']]
idf.index=np.arange(1,11)
idf
```

```python
#Interpetation :-According to  analysis, maximum of Importation of girls has been reported in Bihar in the year 2011
```

```python
#Let us have a look on how more cases from each state started coming up more frequently with each passing year.
count_df = crime_df.groupby('Year')[['STATE/UT']].count()
count_df
```

```python
# Let us use seaborn to draw the heatmap 
plt.figure(figsize=(10,5))
plt.ylabel('Year',fontsize=15)
plt.title("How frequently cases have been reported from every state Year wise",bbox={'facecolor':'0.9','pad':5},pad=20)
sns.heatmap(count_df,cmap='copper');
```

```python
plt.figure(figsize=(10,5))
plt.ylabel('Year',fontsize=15)
plt.title("How frequently cases have been reported from every state Year wise")
sns.heatmap(count_df)
```

```python
#this heatmap shows how states' names started appearing more frequently on the dataset, for reporting crimes and sadly the crime only increased with the passing year.

#It gives us the conclusion that overall(cases under each category), from every state, has increased with time, not decreased.
```
